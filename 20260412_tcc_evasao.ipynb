#Bloco 4: Carregar os dados (comeÃ§ando com 2023)

```python
# Carregar dados de CURSOS 2023
print("Carregando MICRODADOS_CADASTRO_CURSOS_2023.CSV...")
df_cursos_2023 = pd.read_csv(
    caminho_base + arquivos['cursos_2023'],
    sep=';',
    encoding='latin1',
    low_memory=False
)
print(f"Shape: {df_cursos_2023.shape}")
print(f"Colunas: {df_cursos_2023.columns.tolist()[:10]}... (total de {len(df_cursos_2023.columns)} colunas)")

# Carregar dados de IES 2023
print("\nCarregando MICRODADOS_ED_SUP_IES_2023.CSV...")
df_ies_2023 = pd.read_csv(
    caminho_base + arquivos['ies_2023'],
    sep=';',
    encoding='latin1',
    low_memory=False
)
print(f"Shape: {df_ies_2023.shape}")

print("\nâ Dados de 2023 carregados com sucesso!")
```

#Bloco 5: Filtrar apenas cursos EaD e criar target de evasÃ£o

```python
# Filtrar apenas cursos EaD (TP_MODALIDADE_ENSINO = 2)
df_ead_2023 = df_cursos_2023[df_cursos_2023['TP_MODALIDADE_ENSINO'] == 2].copy()
print(f"Total de registros (todos os cursos): {len(df_cursos_2023)}")
print(f"Total de registros (apenas EaD): {len(df_ead_2023)}")
print(f"Percentual de cursos EaD: {len(df_ead_2023)/len(df_cursos_2023)*100:.2f}%")

# Criar variÃ¡vel target: Taxa de EvasÃ£o
# Desvinculados = evasÃ£o
df_ead_2023['QT_EVASAO'] = df_ead_2023['QT_SIT_DESVINCULADO'].fillna(0)

# Opcional: incluir trancados como evasÃ£o (podemos testar depois)
# df_ead_2023['QT_EVASAO_AMPLIADA'] = df_ead_2023['QT_SIT_DESVINCULADO'].fillna(0) + df_ead_2023['QT_SIT_TRANCADA'].fillna(0)

# Calcular taxa de evasÃ£o (evasÃ£o / ingressantes)
# Protegendo contra divisÃ£o por zero
df_ead_2023['QT_ING'] = df_ead_2023['QT_ING'].fillna(0)
df_ead_2023['TAXA_EVASAO'] = df_ead_2023.apply(
    lambda row: row['QT_EVASAO'] / row['QT_ING'] if row['QT_ING'] > 0 else np.nan,
    axis=1
)

print("\nâ Target criado com sucesso!")
print(f"MÃ©dia da taxa de evasÃ£o: {df_ead_2023['TAXA_EVASAO'].mean():.2%}")
print(f"Mediana da taxa de evasÃ£o: {df_ead_2023['TAXA_EVASAO'].median():.2%}")
```

#Bloco 6: VisualizaÃ§Ã£o inicial da distribuiÃ§Ã£o da evasÃ£o

```python
# DistribuiÃ§Ã£o da taxa de evasÃ£o
plt.figure(figsize=(12, 5))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(df_ead_2023['TAXA_EVASAO'].dropna(), bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Taxa de EvasÃ£o')
plt.ylabel('NÃºmero de Cursos')
plt.title('DistribuiÃ§Ã£o da Taxa de EvasÃ£o nos Cursos EaD (2023)')
plt.axvline(df_ead_2023['TAXA_EVASAO'].mean(), color='red', linestyle='--', label=f'MÃ©dia: {df_ead_2023["TAXA_EVASAO"].mean():.2%}')
plt.legend()

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df_ead_2023['TAXA_EVASAO'].dropna())
plt.ylabel('Taxa de EvasÃ£o')
plt.title('Boxplot da Taxa de EvasÃ£o')

plt.tight_layout()
plt.show()

print(f"EstatÃ­sticas descritivas da taxa de evasÃ£o:")
print(df_ead_2023['TAXA_EVASAO'].describe())
```

#Bloco 11 (AnÃ¡lise por RegiÃ£o)

```python
# ==========================================
# ANÃLISE 2023 - EVASÃO POR REGIÃO (CORRIGIDO)
# ==========================================

print("="*50)
print("EVASÃO POR REGIÃO - 2023")
print("="*50)

# Usar NO_REGIAO_IES (sem _y)
evasao_regiao_2023 = df_completo_2023.groupby('NO_REGIAO_IES')['TAXA_EVASAO'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)
print(evasao_regiao_2023)

# ==========================================
# ANÃLISE 2024 - EVASÃO POR REGIÃO (CORRIGIDO)
# ==========================================

print("\n" + "="*50)
print("EVASÃO POR REGIÃO - 2024")
print("="*50)

evasao_regiao_2024 = df_completo_2024.groupby('NO_REGIAO_IES')['TAXA_EVASAO'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)
print(evasao_regiao_2024)

# ==========================================
# GRÃFICO COMPARATIVO
# ==========================================

# Verificar quais regiÃµes existem em ambos os anos
regioes_2023 = set(evasao_regiao_2023.index)
regioes_2024 = set(evasao_regiao_2024.index)
regioes_comuns = sorted(regioes_2023.intersection(regioes_2024))

print("\nRegiÃµes comuns a ambos os anos:", regioes_comuns)

# Preparar dados para o grÃ¡fico (apenas regiÃµes comuns)
medias_2023 = [evasao_regiao_2023.loc[r]['mean'] for r in regioes_comuns]
medias_2024 = [evasao_regiao_2024.loc[r]['mean'] for r in regioes_comuns]

x = range(len(regioes_comuns))
width = 0.35

plt.figure(figsize=(12, 6))
plt.bar([i - width/2 for i in x], medias_2023, width, label='2023', alpha=0.8, color='steelblue')
plt.bar([i + width/2 for i in x], medias_2024, width, label='2024', alpha=0.8, color='darkorange')
plt.xlabel('RegiÃ£o')
plt.ylabel('Taxa MÃ©dia de EvasÃ£o')
plt.title('ComparaÃ§Ã£o da Taxa de EvasÃ£o por RegiÃ£o (2023 vs 2024)')
plt.xticks(x, regioes_comuns, rotation=45)
plt.legend()
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# ==========================================
# TABELA COMPARATIVA
# ==========================================

print("\n" + "="*50)
print("TABELA COMPARATIVA DE EVASÃO POR REGIÃO")
print("="*50)

comparacao_regiao = pd.DataFrame({
    '2023 (mÃ©dia)': medias_2023,
    '2024 (mÃ©dia)': medias_2024,
    'DiferenÃ§a (p.p.)': [(medias_2024[i] - medias_2023[i])*100 for i in range(len(regioes_comuns))]
}, index=regioes_comuns)

print(comparacao_regiao.sort_values('2024 (mÃ©dia)', ascending=False))
```

#Bloco 12 (AnÃ¡lise por OrganizaÃ§Ã£o AcadÃªmica)

```python
# ==========================================
# ANÃLISE 2023 - EVASÃO POR ORGANIZAÃÃO ACADÃMICA (CORRIGIDO)
# ==========================================

print("="*50)
print("EVASÃO POR ORGANIZAÃÃO ACADÃMICA - 2023")
print("="*50)

# Usar TP_ORGANIZACAO_ACADEMICA_y (vem do arquivo IES)
# Legenda: 1=Universidade, 2=Centro UniversitÃ¡rio, 3=Faculdade, 4=Instituto Federal, 5=Centro Federal
evasao_org_2023 = df_completo_2023.groupby('TP_ORGANIZACAO_ACADEMICA_y')['TAXA_EVASAO'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)
print(evasao_org_2023)

# ==========================================
# ANÃLISE 2024 - EVASÃO POR ORGANIZAÃÃO ACADÃMICA (CORRIGIDO)
# ==========================================

print("\n" + "="*50)
print("EVASÃO POR ORGANIZAÃÃO ACADÃMICA - 2024")
print("="*50)

evasao_org_2024 = df_completo_2024.groupby('TP_ORGANIZACAO_ACADEMICA_y')['TAXA_EVASAO'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)
print(evasao_org_2024)

# ==========================================
# GRÃFICOS
# ==========================================

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df_completo_2023, x='TP_ORGANIZACAO_ACADEMICA_y', y='TAXA_EVASAO')
plt.title('Taxa de EvasÃ£o por OrganizaÃ§Ã£o AcadÃªmica (2023)')
plt.xlabel('OrganizaÃ§Ã£o AcadÃªmica (1=Univ, 2=CentroUniv, 3=Fac, 4=IF, 5=CEFET)')
plt.ylabel('Taxa de EvasÃ£o')
plt.ylim(0, 1)

plt.subplot(1, 2, 2)
sns.boxplot(data=df_completo_2024, x='TP_ORGANIZACAO_ACADEMICA_y', y='TAXA_EVASAO')
plt.title('Taxa de EvasÃ£o por OrganizaÃ§Ã£o AcadÃªmica (2024)')
plt.xlabel('OrganizaÃ§Ã£o AcadÃªmica (1=Univ, 2=CentroUniv, 3=Fac, 4=IF, 5=CEFET)')
plt.ylabel('Taxa de EvasÃ£o')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()

# ==========================================
# TABELA COMPARATIVA
# ==========================================

print("\n" + "="*50)
print("TABELA COMPARATIVA - ORGANIZAÃÃO ACADÃMICA")
print("="*50)

# Mapeamento dos cÃ³digos
org_map = {1: 'Universidade', 2: 'Centro UniversitÃ¡rio', 3: 'Faculdade', 4: 'Instituto Federal', 5: 'Centro Federal'}

# Criar tabela comparativa
comparacao_org = []

for cod in evasao_org_2023.index:
    tipo = org_map.get(cod, str(cod))
    media_2023 = evasao_org_2023.loc[cod, 'mean']
    count_2023 = evasao_org_2023.loc[cod, 'count']

    if cod in evasao_org_2024.index:
        media_2024 = evasao_org_2024.loc[cod, 'mean']
        count_2024 = evasao_org_2024.loc[cod, 'count']
        variacao = (media_2024 - media_2023) * 100
    else:
        media_2024 = None
        count_2024 = 0
        variacao = None

    comparacao_org.append({
        'Tipo': tipo,
        '2023 (mÃ©dia)': media_2023,
        '2024 (mÃ©dia)': media_2024,
        'Count 2023': count_2023,
        'Count 2024': count_2024,
        'VariaÃ§Ã£o (p.p.)': variacao
    })

df_comparacao_org = pd.DataFrame(comparacao_org)
print(df_comparacao_org.to_string(index=False))
```

# Bloco 16 PreparacÃ£o dos Daos

## 1.PreparaÃ§Ã£o dos dados

```python
# ==========================================
# 1. SELECIONAR AS VARIÃVEIS RELEVANTES
# ==========================================

# Vamos usar df_completo_2023 como base (jÃ¡ limpo e com merge)
# Identificar as colunas que serÃ£o usadas como features (variÃ¡veis preditoras)

# Lista de variÃ¡veis candidatas (com base no dicionÃ¡rio e anÃ¡lises anteriores)
features_candidatas = [
    # VariÃ¡veis demogrÃ¡ficas (proporÃ§Ãµes)
    'PROP_FEM',                    # ProporÃ§Ã£o de mulheres
    'PROP_ESCOLA_PUBLICA',         # ProporÃ§Ã£o de alunos de escola pÃºblica

    # VariÃ¡veis de apoio financeiro (proporÃ§Ãµes)
    'PROP_FIES',                   # ProporÃ§Ã£o de alunos com FIES
    'PROP_PROUNI',                 # ProporÃ§Ã£o de alunos com PROUNI

    # VariÃ¡veis institucionais
    'TP_REDE_y',                   # Rede: 1=PÃºblica, 2=Privada
    'TP_ORGANIZACAO_ACADEMICA_y',  # Tipo de organizaÃ§Ã£o acadÃªmica
    'NO_REGIAO_IES',               # RegiÃ£o geogrÃ¡fica

    # VariÃ¡veis do curso
    'TP_GRAU_ACADEMICO',           # Grau: Bacharelado, Licenciatura, TecnolÃ³gico
]

# Verificar quais colunas realmente existem no dataframe
features_disponiveis = [f for f in features_candidatas if f in df_completo_2023.columns]
print(f"Features disponÃ­veis: {len(features_disponiveis)} de {len(features_candidatas)}")
for f in features_disponiveis:
    print(f"  - {f}")
```

## 2.Dataframe Modelagem

```python
# ==========================================
# 2. CRIAR DATAFRAME DE MODELAGEM
# ==========================================

# Selecionar as features disponÃ­veis e o target
df_model = df_completo_2023[features_disponiveis + ['TAXA_EVASAO']].copy()

# Remover linhas com target ausente (TAXA_EVASAO)
df_model = df_model.dropna(subset=['TAXA_EVASAO'])
print(f"\nShape apÃ³s remover target ausente: {df_model.shape}")

# Remover linhas com valores ausentes em features (opcional - podemos analisar)
print(f"Valores ausentes por coluna:")
print(df_model.isnull().sum())
```

## 3.Tratar missing

```python
# ==========================================
# 3. TRATAR VALORES AUSENTES (CORRIGIDO - SEM WARNINGS)
# ==========================================

# Para variÃ¡veis numÃ©ricas: preencher com a mediana
variaveis_numericas = ['PROP_FEM', 'PROP_ESCOLA_PUBLICA', 'PROP_FIES', 'PROP_PROUNI']
for var in variaveis_numericas:
    if var in df_model.columns:
        mediana = df_model[var].median()
        df_model[var] = df_model[var].fillna(mediana)  # Sintaxe correta
        print(f"Preenchido {var} com mediana: {mediana:.4f}")

# Para variÃ¡veis categÃ³ricas: preencher com a moda (valor mais frequente)
variaveis_categoricas = ['TP_REDE_y', 'TP_ORGANIZACAO_ACADEMICA_y', 'NO_REGIAO_IES', 'TP_GRAU_ACADEMICO']
for var in variaveis_categoricas:
    if var in df_model.columns:
        moda = df_model[var].mode()[0]
        df_model[var] = df_model[var].fillna(moda)  # Sintaxe correta
        print(f"Preenchido {var} com moda: {moda}")

print("\nâ Valores ausentes tratados!")
```

## 4.VariÃ¡veis categÃ³ricas

```python
# ==========================================
# 4. CODIFICAR VARIÃVEIS CATEGÃRICAS
# ==========================================

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# 4.1 Para variÃ¡veis ordinais (Label Encoding)
# TP_GRAU_ACADEMICO: 1=Bacharelado, 2=Licenciatura, 3=TecnolÃ³gico
if 'TP_GRAU_ACADEMICO' in df_model.columns:
    le_grau = LabelEncoder()
    df_model['TP_GRAU_ACADEMICO_ENCODED'] = le_grau.fit_transform(df_model['TP_GRAU_ACADEMICO'].astype(str))
    print(f"\nLabel Encoding - TP_GRAU_ACADEMICO: {dict(zip(le_grau.classes_, le_grau.transform(le_grau.classes_)))}")

# 4.2 Para variÃ¡veis nominais (One-Hot Encoding)
# NO_REGIAO_IES - RegiÃ£o
if 'NO_REGIAO_IES' in df_model.columns:
    dummies_regiao = pd.get_dummies(df_model['NO_REGIAO_IES'], prefix='REGIAO', drop_first=True)
    df_model = pd.concat([df_model, dummies_regiao], axis=1)
    print(f"\nOne-Hot Encoding - REGIAO: {dummies_regiao.columns.tolist()}")

# TP_ORGANIZACAO_ACADEMICA_y
if 'TP_ORGANIZACAO_ACADEMICA_y' in df_model.columns:
    # Mapeamento para nomes mais descritivos
    org_map = {1: 'UNIVERSIDADE', 2: 'CENTRO_UNIVERSITARIO', 3: 'FACULDADE', 4: 'INSTITUTO_FEDERAL', 5: 'CENTRO_FEDERAL'}
    df_model['ORGANIZACAO_NOME'] = df_model['TP_ORGANIZACAO_ACADEMICA_y'].map(org_map)
    dummies_org = pd.get_dummies(df_model['ORGANIZACAO_NOME'], prefix='ORG', drop_first=True)
    df_model = pd.concat([df_model, dummies_org], axis=1)
    print(f"One-Hot Encoding - ORGANIZACAO: {dummies_org.columns.tolist()}")

# TP_REDE_y (jÃ¡ Ã© binÃ¡rio, mas vamos garantir)
if 'TP_REDE_y' in df_model.columns:
    df_model['TP_REDE_y'] = df_model['TP_REDE_y'].astype(int)

print("\nâ VariÃ¡veis categÃ³ricas codificadas!")
```

## 5.SeleÃ§Ã£o de features

```python
# ==========================================
# 5. SELECIONAR FEATURES FINAIS PARA MODELAGEM
# ==========================================

# Lista das features que serÃ£o usadas no modelo
features_finais = []

# VariÃ¡veis numÃ©ricas originais
for var in variaveis_numericas:
    if var in df_model.columns:
        features_finais.append(var)

# VariÃ¡veis codificadas
if 'TP_GRAU_ACADEMICO_ENCODED' in df_model.columns:
    features_finais.append('TP_GRAU_ACADEMICO_ENCODED')

# Dummies de regiÃ£o
for col in dummies_regiao.columns:
    features_finais.append(col)

# Dummies de organizaÃ§Ã£o
for col in dummies_org.columns:
    features_finais.append(col)

# Rede
if 'TP_REDE_y' in df_model.columns:
    features_finais.append('TP_REDE_y')

print(f"\nFeatures finais para modelagem: {len(features_finais)}")
print(features_finais)


```

# Bloco 17 - ExecuÃ§Ã£o dos Modelos (Treino e teste)

```python
# ==========================================
# BLOCO 17 CORRIGIDO: MODELAGEM PREDITIVA
# ==========================================

print("="*60)
print("MODELAGEM PREDITIVA - PREVISÃO DE EVASÃO")
print("="*60)

# ==========================================
# 1. IMPORTAR BIBLIOTECAS NECESSÃRIAS
# ==========================================

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix,
                             roc_curve)  # <-- CORREÃÃO: adicionado roc_curve
import warnings
warnings.filterwarnings('ignore')

print("â Bibliotecas importadas!")

# ==========================================
# 2. VERIFICAR SE OS DADOS ESTÃO PRONTOS
# ==========================================

# Garantir que as variÃ¡veis do prepared_data existam
if 'prepared_data' not in locals():
    print("â ï¸ prepared_data nÃ£o encontrado. Execute o Bloco 16 primeiro!")
else:
    X_train = prepared_data['X_train_scaled']
    X_test = prepared_data['X_test_scaled']
    y_train = prepared_data['y_train_bin']
    y_test = prepared_data['y_test_bin']

    print(f"Dados carregados:")
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_train - Classe 1 (alta evasÃ£o): {y_train.mean():.2%}")
    print(f"y_test - Classe 1 (alta evasÃ£o): {y_test.mean():.2%}")

# ==========================================
# 3. DEFINIR OS MODELOS A SEREM TESTADOS
# ==========================================

print("\n" + "="*60)
print("MODELOS A SEREM TREINADOS")
print("="*60)

modelos = {
    'RegressÃ£o LogÃ­stica': LogisticRegression(random_state=42, max_iter=1000),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
}

print("Modelos configurados:")
for nome in modelos.keys():
    print(f"  - {nome}")

# ==========================================
# 4. TREINAR E AVALIAR CADA MODELO
# ==========================================

print("\n" + "="*60)
print("TREINAMENTO E AVALIAÃÃO DOS MODELOS")
print("="*60)

resultados = {}

for nome, modelo in modelos.items():
    print(f"\n--- {nome} ---")

    # Treinar o modelo
    modelo.fit(X_train, y_train)

    # Fazer previsÃµes
    y_pred = modelo.predict(X_test)
    y_pred_proba = modelo.predict_proba(X_test)[:, 1]  # Probabilidade da classe 1

    # Calcular mÃ©tricas
    acuracia = accuracy_score(y_test, y_pred)
    precisao = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred_proba)

    # Armazenar resultados
    resultados[nome] = {
        'AcurÃ¡cia': acuracia,
        'PrecisÃ£o': precisao,
        'Recall': recall,
        'F1-Score': f1,
        'AUC-ROC': auc,
        'modelo': modelo,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba
    }

    # Exibir mÃ©tricas
    print(f"AcurÃ¡cia:  {acuracia:.4f} ({acuracia:.2%})")
    print(f"PrecisÃ£o:  {precisao:.4f} ({precisao:.2%})")
    print(f"Recall:    {recall:.4f} ({recall:.2%})")
    print(f"F1-Score:  {f1:.4f} ({f1:.2%})")
    print(f"AUC-ROC:   {auc:.4f} ({auc:.2%})")

    # Matriz de confusÃ£o
    cm = confusion_matrix(y_test, y_pred)
    print(f"Matriz de ConfusÃ£o:")
    print(f"  Verdadeiros Negativos (VN): {cm[0,0]}")
    print(f"  Falsos Positivos (FP):     {cm[0,1]}")
    print(f"  Falsos Negativos (FN):     {cm[1,0]}")
    print(f"  Verdadeiros Positivos (VP):{cm[1,1]}")

# ==========================================
# 5. COMPARAÃÃO DOS MODELOS
# ==========================================

print("\n" + "="*60)
print("COMPARAÃÃO DOS MODELOS")
print("="*60)

# Criar DataFrame de comparaÃ§Ã£o
df_comparacao = pd.DataFrame(resultados).T
df_comparacao = df_comparacao[['AcurÃ¡cia', 'PrecisÃ£o', 'Recall', 'F1-Score', 'AUC-ROC']]

print("\nTabela Comparativa:")
print(df_comparacao.round(4))

# Identificar o melhor modelo por mÃ©trica
print("\n--- Melhor modelo por mÃ©trica ---")
metricas = ['AcurÃ¡cia', 'PrecisÃ£o', 'Recall', 'F1-Score', 'AUC-ROC']
for metrica in metricas:
    melhor = df_comparacao[metrica].idxmax()
    melhor_valor = df_comparacao[metrica].max()
    print(f"{metrica}: {melhor} ({melhor_valor:.4f})")

# Melhor modelo geral (considerando F1-Score como principal)
melhor_modelo_nome = df_comparacao['F1-Score'].idxmax()
melhor_modelo = resultados[melhor_modelo_nome]['modelo']
print(f"\nð Melhor modelo geral (F1-Score): {melhor_modelo_nome}")

# ==========================================
# 6. GRÃFICO COMPARATIVO
# ==========================================

fig, ax = plt.subplots(figsize=(10, 6))

metricas_plot = ['AcurÃ¡cia', 'PrecisÃ£o', 'Recall', 'F1-Score', 'AUC-ROC']
x = np.arange(len(metricas_plot))
width = 0.25

cores = {'RegressÃ£o LogÃ­stica': 'steelblue', 'Random Forest': 'darkorange', 'XGBoost': 'forestgreen'}

for i, (nome, resultado) in enumerate(resultados.items()):
    valores = [resultado[m] for m in metricas_plot]
    offset = (i - 1) * width
    ax.bar(x + offset, valores, width, label=nome, color=cores[nome], alpha=0.8)

ax.set_xlabel('MÃ©tricas')
ax.set_ylabel('PontuaÃ§Ã£o')
ax.set_title('ComparaÃ§Ã£o de Desempenho dos Modelos')
ax.set_xticks(x)
ax.set_xticklabels(metricas_plot)
ax.legend(loc='lower right')
ax.set_ylim(0, 1)
ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)

for i, (nome, resultado) in enumerate(resultados.items()):
    valores = [resultado[m] for m in metricas_plot]
    offset = (i - 1) * width
    for j, v in enumerate(valores):
        ax.text(j + offset, v + 0.02, f'{v:.2f}', ha='center', fontsize=8)

plt.tight_layout()
plt.show()

# ==========================================
# 7. MATRIZES DE CONFUSÃO
# ==========================================

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, (nome, resultado) in enumerate(resultados.items()):
    cm = confusion_matrix(y_test, resultado['y_pred'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx])
    axes[idx].set_title(f'{nome}')
    axes[idx].set_xlabel('Predito')
    axes[idx].set_ylabel('Real')
    axes[idx].set_xticklabels(['Baixa EvasÃ£o', 'Alta EvasÃ£o'])
    axes[idx].set_yticklabels(['Baixa EvasÃ£o', 'Alta EvasÃ£o'])

plt.tight_layout()
plt.show()

# ==========================================
# 8. CURVAS ROC (CORRIGIDO)
# ==========================================

from sklearn.metrics import roc_curve  # Garantindo a importaÃ§Ã£o

plt.figure(figsize=(10, 6))

for nome, resultado in resultados.items():
    fpr, tpr, _ = roc_curve(y_test, resultado['y_pred_proba'])
    auc = resultado['AUC-ROC']
    plt.plot(fpr, tpr, label=f'{nome} (AUC = {auc:.3f})', linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', label='AleatÃ³rio (AUC = 0.5)')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curvas ROC dos Modelos')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# ==========================================
# 9. IMPORTÃNCIA DAS FEATURES (Random Forest)
# ==========================================

print("\n" + "="*60)
print("IMPORTÃNCIA DAS FEATURES - RANDOM FOREST")
print("="*60)

if 'Random Forest' in resultados:
    rf_model = resultados['Random Forest']['modelo']
    importancias = rf_model.feature_importances_

    # Criar DataFrame com importÃ¢ncias
    df_importancia = pd.DataFrame({
        'Feature': prepared_data['features_finais'],
        'ImportÃ¢ncia': importancias
    }).sort_values('ImportÃ¢ncia', ascending=False)

    print("\nTop 10 features mais importantes:")
    print(df_importancia.head(10).to_string(index=False))

    # GrÃ¡fico de importÃ¢ncias
    plt.figure(figsize=(10, 6))
    top_features = df_importancia.head(10)
    plt.barh(top_features['Feature'], top_features['ImportÃ¢ncia'], color='steelblue')
    plt.xlabel('ImportÃ¢ncia')
    plt.title('Top 10 Features Mais Importantes (Random Forest)')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

# ==========================================
# 10. SALVAR RESULTADOS
# ==========================================

modelagem_resultados = {
    'resultados': resultados,
    'melhor_modelo_nome': melhor_modelo_nome,
    'melhor_modelo': melhor_modelo,
    'df_comparacao': df_comparacao,
    'df_importancia': df_importancia if 'df_importancia' in locals() else None
}

print("\nâ Resultados da modelagem salvos em 'modelagem_resultados'!")

# ==========================================
# 11. RESUMO FINAL
# ==========================================

print("\n" + "="*60)
print("RESUMO DA MODELAGEM PREDITIVA")
print("="*60)
print(f"Total de amostras: {len(X_train) + len(X_test)}")
print(f"Treino: {len(X_train)} (70%)")
print(f"Teste: {len(X_test)} (30%)")
print(f"Features utilizadas: {len(prepared_data['features_finais'])}")
print(f"\nð Melhor modelo: {melhor_modelo_nome}")
print(f"   AcurÃ¡cia: {resultados[melhor_modelo_nome]['AcurÃ¡cia']:.4f} ({resultados[melhor_modelo_nome]['AcurÃ¡cia']:.2%})")
print(f"   PrecisÃ£o: {resultados[melhor_modelo_nome]['PrecisÃ£o']:.4f} ({resultados[melhor_modelo_nome]['PrecisÃ£o']:.2%})")
print(f"   Recall:   {resultados[melhor_modelo_nome]['Recall']:.4f} ({resultados[melhor_modelo_nome]['Recall']:.2%})")
print(f"   F1-Score: {resultados[melhor_modelo_nome]['F1-Score']:.4f} ({resultados[melhor_modelo_nome]['F1-Score']:.2%})")
print(f"   AUC-ROC:  {resultados[melhor_modelo_nome]['AUC-ROC']:.4f} ({resultados[melhor_modelo_nome]['AUC-ROC']:.2%})")

print("\nâ Modelagem concluÃ­da com sucesso!")
```

# Bloco 18 - AvaliaÃ§Ã£o Detalhada do Melhor Modelo (XGBoost)

```python
# ==========================================
# BLOCO 18: AVALIAÃÃO DETALHADA DO MELHOR MODELO (XGBoost)
# ==========================================

print("="*60)
print("AVALIAÃÃO DETALHADA DO MELHOR MODELO - XGBoost")
print("="*60)

# ==========================================
# 1. IMPORTAR BIBLIOTECAS ADICIONAIS
# ==========================================

from sklearn.model_selection import learning_curve, cross_val_score, StratifiedKFold
from sklearn.metrics import roc_curve, auc, classification_report
import warnings
warnings.filterwarnings('ignore')

print("â Bibliotecas importadas!")

# ==========================================
# 2. CARREGAR O MELHOR MODELO
# ==========================================

# Recuperar o melhor modelo (XGBoost) dos resultados
melhor_modelo = resultados['XGBoost']['modelo']
melhor_nome = 'XGBoost'

print(f"Melhor modelo: {melhor_nome}")
print(f"Modelo carregado: {type(melhor_modelo).__name__}")

# ==========================================
# 3. CURVA DE APRENDIZADO (Learning Curve)
# ==========================================

print("\n" + "="*60)
print("CURVA DE APRENDIZADO")
print("="*60)

# Calcular curva de aprendizado
train_sizes, train_scores, test_scores = learning_curve(
    melhor_modelo, X_train, y_train,
    cv=5,  # 5-fold cross-validation
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='f1',
    random_state=42
)

# Calcular mÃ©dias e desvios
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

print(f"Treinamento: {len(train_sizes)} tamanhos de amostra testados")
print(f"Menor amostra: {train_sizes[0]:.0f} registros")
print(f"Maior amostra: {train_sizes[-1]:.0f} registros")

# GrÃ¡fico da Curva de Aprendizado
plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='steelblue')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='darkorange')
plt.plot(train_sizes, train_mean, 'o-', color='steelblue', label='Treino (F1-Score)')
plt.plot(train_sizes, test_mean, 'o-', color='darkorange', label='ValidaÃ§Ã£o (F1-Score)')
plt.xlabel('Tamanho do Conjunto de Treino')
plt.ylabel('F1-Score')
plt.title('Curva de Aprendizado - XGBoost')
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# InterpretaÃ§Ã£o da curva de aprendizado
if test_mean[-1] > test_mean[0]:
    print("\nâ A curva de aprendizado estÃ¡ subindo com mais dados -> O modelo pode melhorar com mais amostras")
else:
    print("\nâ ï¸ A curva de aprendizado estabilizou -> Mais dados podem nÃ£o melhorar significativamente")

if train_mean[-1] - test_mean[-1] > 0.1:
    print(f"â ï¸ PossÃ­vel overfitting: diferenÃ§a de {(train_mean[-1] - test_mean[-1]):.4f} entre treino e validaÃ§Ã£o")
else:
    print(f"â Modelo bem ajustado: diferenÃ§a de {(train_mean[-1] - test_mean[-1]):.4f} entre treino e validaÃ§Ã£o")

# ==========================================
# 4. VALIDAÃÃO CRUZADA (CROSS-VALIDATION)
# ==========================================

print("\n" + "="*60)
print("VALIDAÃÃO CRUZADA (5-FOLD)")
print("="*60)

# ValidaÃ§Ã£o cruzada com 5 folds
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_f1 = cross_val_score(melhor_modelo, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)
cv_scores_accuracy = cross_val_score(melhor_modelo, X_train, y_train, cv=cv, scoring='accuracy', n_jobs=-1)
cv_scores_roc_auc = cross_val_score(melhor_modelo, X_train, y_train, cv=cv, scoring='roc_auc', n_jobs=-1)

print("\nResultados da ValidaÃ§Ã£o Cruzada (5-fold):")
print(f"F1-Score:     mÃ©dia = {cv_scores_f1.mean():.4f} (Â±{cv_scores_f1.std():.4f})")
print(f"  Folds: {[f'{x:.4f}' for x in cv_scores_f1]}")
print(f"AcurÃ¡cia:     mÃ©dia = {cv_scores_accuracy.mean():.4f} (Â±{cv_scores_accuracy.std():.4f})")
print(f"  Folds: {[f'{x:.4f}' for x in cv_scores_accuracy]}")
print(f"AUC-ROC:      mÃ©dia = {cv_scores_roc_auc.mean():.4f} (Â±{cv_scores_roc_auc.std():.4f})")
print(f"  Folds: {[f'{x:.4f}' for x in cv_scores_roc_auc]}")

# GrÃ¡fico dos scores por fold
plt.figure(figsize=(10, 5))
folds = range(1, 6)
plt.plot(folds, cv_scores_f1, 'o-', label='F1-Score', color='steelblue', marker='o')
plt.plot(folds, cv_scores_accuracy, 's-', label='AcurÃ¡cia', color='darkorange', marker='s')
plt.plot(folds, cv_scores_roc_auc, '^-', label='AUC-ROC', color='forestgreen', marker='^')
plt.axhline(y=cv_scores_f1.mean(), color='steelblue', linestyle='--', alpha=0.5)
plt.axhline(y=cv_scores_accuracy.mean(), color='darkorange', linestyle='--', alpha=0.5)
plt.axhline(y=cv_scores_roc_auc.mean(), color='forestgreen', linestyle='--', alpha=0.5)
plt.xlabel('Fold')
plt.ylabel('Score')
plt.title('ValidaÃ§Ã£o Cruzada (5-fold) - XGBoost')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# ==========================================
# 5. CURVA ROC DETALHADA COM INTERVALO DE CONFIANÃA
# ==========================================

print("\n" + "="*60)
print("CURVA ROC COM INTERVALO DE CONFIANÃA")
print("="*60)

# Calcular ROC para o conjunto de teste
y_pred_proba = resultados['XGBoost']['y_pred_proba']
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Calcular intervalo de confianÃ§a (bootstrap simples)
n_bootstraps = 1000
auc_scores = []

rng = np.random.RandomState(42)
for _ in range(n_bootstraps):
    indices = rng.randint(0, len(y_test), len(y_test))
    if len(np.unique(y_test.iloc[indices])) < 2:
        continue
    auc_scores.append(roc_auc_score(y_test.iloc[indices], y_pred_proba[indices]))

auc_lower = np.percentile(auc_scores, 2.5)
auc_upper = np.percentile(auc_scores, 97.5)

print(f"AUC-ROC: {roc_auc:.4f}")
print(f"Intervalo de ConfianÃ§a 95%: [{auc_lower:.4f} - {auc_upper:.4f}]")

# GrÃ¡fico ROC com intervalo
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC (AUC = {roc_auc:.3f})')
plt.fill_between(fpr,
                 np.maximum(0, tpr - 0.05),
                 np.minimum(1, tpr + 0.05),
                 alpha=0.2, color='darkorange', label='IC 95% (aproximado)')
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--', label='AleatÃ³rio (AUC = 0.5)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC - XGBoost (com Intervalo de ConfianÃ§a)')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# ==========================================
# 6. ANÃLISE DE ERROS (CLASSIFICAÃÃO INCORRETA)
# ==========================================

print("\n" + "="*60)
print("ANÃLISE DE ERROS")
print("="*60)

# PrevisÃµes do melhor modelo
y_pred = resultados['XGBoost']['y_pred']

# Identificar onde o modelo errou
erros_fp = (y_pred == 1) & (y_test == 0)  # Falsos Positivos (previu evasÃ£o, mas nÃ£o houve)
erros_fn = (y_pred == 0) & (y_test == 1)  # Falsos Negativos (nÃ£o previu evasÃ£o, mas houve)

print(f"Total de previsÃµes: {len(y_test)}")
print(f"Acertos: {(y_pred == y_test).sum()} ({(y_pred == y_test).mean():.2%})")
print(f"Erros: {(y_pred != y_test).sum()} ({(y_pred != y_test).mean():.2%})")
print(f"  - Falsos Positivos (FP - alarme falso): {erros_fp.sum()} ({erros_fp.mean():.2%})")
print(f"  - Falsos Negativos (FN - evasÃ£o nÃ£o detectada): {erros_fn.sum()} ({erros_fn.mean():.2%})")

# ==========================================
# 7. RELATÃRIO DE CLASSIFICAÃÃO DETALHADO
# ==========================================

print("\n" + "="*60)
print("RELATÃRIO DE CLASSIFICAÃÃO DETALHADO")
print("="*60)

print("\nClassification Report:")
print(classification_report(y_test, y_pred,
                            target_names=['Baixa EvasÃ£o', 'Alta EvasÃ£o'],
                            digits=4))

# ==========================================
# 8. MÃTRICAS ADICIONAIS
# ==========================================

print("\n" + "="*60)
print("MÃTRICAS ADICIONAIS")
print("="*60)

from sklearn.metrics import matthews_corrcoef, balanced_accuracy_score

mcc = matthews_corrcoef(y_test, y_pred)
balanced_acc = balanced_accuracy_score(y_test, y_pred)

print(f"Coeficiente de Matthews (MCC): {mcc:.4f}")
print(f"  (1 = perfeito, 0 = aleatÃ³rio, -1 = inverso)")
print(f"AcurÃ¡cia Balanceada: {balanced_acc:.4f} ({balanced_acc:.2%})")

# ==========================================
# 9. INTERPRETAÃÃO DOS RESULTADOS
# ==========================================

print("\n" + "="*60)
print("INTERPRETAÃÃO DOS RESULTADOS")
print("="*60)

# AvaliaÃ§Ã£o geral
print("\n--- AvaliaÃ§Ã£o do Modelo ---")
if roc_auc > 0.7:
    print("â O modelo tem boa capacidade de discriminaÃ§Ã£o (AUC-ROC > 0.7)")
elif roc_auc > 0.6:
    print("â ï¸ O modelo tem capacidade razoÃ¡vel de discriminaÃ§Ã£o (AUC-ROC entre 0.6 e 0.7)")
else:
    print("â O modelo tem baixa capacidade de discriminaÃ§Ã£o (AUC-ROC < 0.6)")

# Overfitting
train_f1 = resultados['XGBoost']['F1-Score']
if train_f1 - cv_scores_f1.mean() > 0.05:
    print(f"â ï¸ PossÃ­vel overfitting: treino F1={train_f1:.4f} vs CV={cv_scores_f1.mean():.4f}")
else:
    print(f"â Modelo generaliza bem: treino F1={train_f1:.4f} vs CV={cv_scores_f1.mean():.4f}")

# Estabilidade
if cv_scores_f1.std() < 0.01:
    print("â Modelo estÃ¡vel: baixa variaÃ§Ã£o entre folds na validaÃ§Ã£o cruzada")
else:
    print(f"â ï¸ Modelo com alguma instabilidade: desvio padrÃ£o de {cv_scores_f1.std():.4f}")

# ==========================================
# 10. SALVAR RESULTADOS DA AVALIAÃÃO
# ==========================================

avaliacao_detalhada = {
    'learning_curve': {
        'train_sizes': train_sizes,
        'train_scores': train_scores,
        'test_scores': test_scores,
        'train_mean': train_mean,
        'test_mean': test_mean
    },
    'cross_validation': {
        'f1_scores': cv_scores_f1,
        'accuracy_scores': cv_scores_accuracy,
        'roc_auc_scores': cv_scores_roc_auc,
        'f1_mean': cv_scores_f1.mean(),
        'f1_std': cv_scores_f1.std()
    },
    'roc': {
        'fpr': fpr,
        'tpr': tpr,
        'auc': roc_auc,
        'auc_ci_lower': auc_lower,
        'auc_ci_upper': auc_upper
    },
    'erros': {
        'falsos_positivos': erros_fp.sum(),
        'falsos_negativos': erros_fn.sum(),
        'total_erros': (y_pred != y_test).sum()
    },
    'metricas_adicionais': {
        'mcc': mcc,
        'balanced_accuracy': balanced_acc
    }
}

print("\nâ AvaliaÃ§Ã£o detalhada salva em 'avaliacao_detalhada'!")

# ==========================================
# 11. RESUMO FINAL DA AVALIAÃÃO
# ==========================================

print("\n" + "="*60)
print("RESUMO DA AVALIAÃÃO DO XGBoost")
print("="*60)
print(f"AUC-ROC: {roc_auc:.4f} (IC 95%: [{auc_lower:.4f} - {auc_upper:.4f}])")
print(f"F1-Score (teste): {resultados['XGBoost']['F1-Score']:.4f}")
print(f"F1-Score (CV mÃ©dia): {cv_scores_f1.mean():.4f} (Â±{cv_scores_f1.std():.4f})")
print(f"MCC: {mcc:.4f}")
print(f"AcurÃ¡cia Balanceada: {balanced_acc:.4f}")

print("\nâ AvaliaÃ§Ã£o detalhada concluÃ­da!")
```

```python
# =============================================================================
# SEÃÃO 2 â GRÃFICOS PLOTLY (VERSÃO CORRIGIDA - SEM ERRO DE TITLE)
# =============================================================================

import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Cores personalizadas
COR_2023 = '#1f77b4'      # azul
COR_2024 = '#ff7f0e'      # laranja
COR_PUBLICA = '#2ca02c'   # verde
COR_PRIVADA = '#d62728'   # vermelho

# ConfiguraÃ§Ã£o padrÃ£o do layout (sem o title para evitar conflito)
LAYOUT_PADRAO = {
    "plot_bgcolor": "white",
    "paper_bgcolor": "white",
    "font": {"family": "Arial, sans-serif", "size": 12},
    "hovermode": "x unified",
    "legend": {"orientation": "h", "yanchor": "bottom", "y": 1.02, "xanchor": "right", "x": 1},
}

# -----------------------------------------------------------------------
# GRÃFICO 1: EvasÃ£o por Rede de Ensino (PÃºblica vs Privada) â 2023 x 2024
# -----------------------------------------------------------------------
def grafico_rede():
    dados = {
        "Rede": ["PÃºblica", "Privada", "PÃºblica", "Privada"],
        "Ano":  ["2023",    "2023",    "2024",    "2024"],
        "Taxa": [
            df_completo_2023[df_completo_2023["TP_REDE_y"] == 1]["TAXA_EVASAO"].mean(),
            df_completo_2023[df_completo_2023["TP_REDE_y"] == 2]["TAXA_EVASAO"].mean(),
            df_completo_2024[df_completo_2024["TP_REDE_y"] == 1]["TAXA_EVASAO"].mean(),
            df_completo_2024[df_completo_2024["TP_REDE_y"] == 2]["TAXA_EVASAO"].mean(),
        ]
    }
    df_plot = pd.DataFrame(dados)

    fig = px.bar(
        df_plot, x="Rede", y="Taxa", color="Ano",
        barmode="group",
        color_discrete_map={"2023": COR_2023, "2024": COR_2024},
        text=df_plot["Taxa"].map(lambda v: f"{v:.1%}"),
        title="Taxa MÃ©dia de EvasÃ£o por Rede de Ensino â 2023 vs 2024",
        labels={"Taxa": "Taxa MÃ©dia de EvasÃ£o", "Rede": ""},
    )
    fig.update_traces(textposition="outside", textfont_size=13)
    fig.update_layout(
        **LAYOUT_PADRAO,
        yaxis_tickformat=".0%",
        yaxis_range=[0, 0.65],
        annotations=[dict(
            text="Cursos EaD | Fonte: Censo INEP 2023-2024",
            xref="paper", yref="paper", x=0, y=-0.12,
            showarrow=False, font=dict(size=11, color="#888888")
        )]
    )
    fig.show()

grafico_rede()


# -----------------------------------------------------------------------
# GRÃFICO 2: EvasÃ£o por RegiÃ£o â Comparativo 2023 x 2024
# -----------------------------------------------------------------------
def grafico_regiao():
    mapa_regiao = {1: "Norte", 2: "Nordeste", 3: "Centro-Oeste",
                   4: "Sudeste", 5: "Sul"}

    def taxa_regiao(df):
        df = df.copy()
        df["NO_REGIAO_NOME"] = df["CO_REGIAO"].map(mapa_regiao)
        return df.groupby("NO_REGIAO_NOME")["TAXA_EVASAO"].mean().reset_index()

    reg_2023 = taxa_regiao(df_completo_2023)
    reg_2023["Ano"] = "2023"
    reg_2024 = taxa_regiao(df_completo_2024)
    reg_2024["Ano"] = "2024"

    df_reg = pd.concat([reg_2023, reg_2024])
    df_reg.columns = ["RegiÃ£o", "Taxa", "Ano"]
    df_reg = df_reg.sort_values(["RegiÃ£o", "Ano"])

    fig = px.bar(
        df_reg, x="RegiÃ£o", y="Taxa", color="Ano",
        barmode="group",
        color_discrete_map={"2023": COR_2023, "2024": COR_2024},
        text=df_reg["Taxa"].map(lambda v: f"{v:.1%}"),
        title="Taxa MÃ©dia de EvasÃ£o por RegiÃ£o â 2023 vs 2024",
        labels={"Taxa": "Taxa MÃ©dia de EvasÃ£o", "RegiÃ£o": ""},
    )
    fig.update_traces(textposition="outside", textfont_size=12)
    fig.update_layout(
        **LAYOUT_PADRAO,
        yaxis_tickformat=".0%",
        yaxis_range=[0, 0.65],
    )
    fig.show()

grafico_regiao()


# -----------------------------------------------------------------------
# GRÃFICO 3: EvasÃ£o por Tipo de OrganizaÃ§Ã£o AcadÃªmica
# -----------------------------------------------------------------------
def grafico_org_academica():
    mapa_org = {
        1: "Universidade",
        2: "Centro UniversitÃ¡rio",
        3: "Faculdade",
        4: "Instituto Federal",
        5: "Centro Federal"
    }

    def taxa_org(df, ano):
        df = df.copy()
        col = "TP_ORGANIZACAO_ACADEMICA_y" if "TP_ORGANIZACAO_ACADEMICA_y" in df.columns else "TP_ORGANIZACAO_ACADEMICA"
        df["ORG_NOME"] = df[col].map(mapa_org)
        result = df.dropna(subset=["ORG_NOME"]) \
                   .groupby("ORG_NOME")["TAXA_EVASAO"].mean().reset_index()
        result.columns = ["OrganizaÃ§Ã£o", "Taxa"]
        result["Ano"] = ano
        return result

    df_org = pd.concat([
        taxa_org(df_completo_2023, "2023"),
        taxa_org(df_completo_2024, "2024")
    ])

    fig = px.bar(
        df_org, x="OrganizaÃ§Ã£o", y="Taxa", color="Ano",
        barmode="group",
        color_discrete_map={"2023": COR_2023, "2024": COR_2024},
        text=df_org["Taxa"].map(lambda v: f"{v:.1%}"),
        title="Taxa MÃ©dia de EvasÃ£o por Tipo de OrganizaÃ§Ã£o AcadÃªmica â 2023 vs 2024",
        labels={"Taxa": "Taxa MÃ©dia de EvasÃ£o", "OrganizaÃ§Ã£o": ""},
    )
    fig.update_traces(textposition="outside", textfont_size=12)
    fig.update_layout(
        **LAYOUT_PADRAO,
        yaxis_tickformat=".0%",
        yaxis_range=[0, 0.75],
    )
    fig.show()

grafico_org_academica()


# -----------------------------------------------------------------------
# GRÃFICO 4: EvasÃ£o por Apoio Financeiro (FIES e ProUni)
# -----------------------------------------------------------------------
def grafico_apoio_financeiro():
    def taxa_apoio(df, ano):
        rows = []
        for col_qt, label in [("QT_ING_FIES", "FIES"), ("QT_ING_PROUNI_TOTAL", "ProUni")]:
            if col_qt in df.columns:
                taxa_com = df[df[col_qt] > 0]["TAXA_EVASAO"].mean()
                taxa_sem = df[df[col_qt] == 0]["TAXA_EVASAO"].mean()
                rows.append({"Programa": label, "SituaÃ§Ã£o": "Com apoio", "Taxa": taxa_com, "Ano": ano})
                rows.append({"Programa": label, "SituaÃ§Ã£o": "Sem apoio", "Taxa": taxa_sem, "Ano": ano})
        return pd.DataFrame(rows)

    df_apoio = pd.concat([
        taxa_apoio(df_completo_2023, "2023"),
        taxa_apoio(df_completo_2024, "2024")
    ])

    fig = px.bar(
        df_apoio, x="Programa", y="Taxa", color="SituaÃ§Ã£o",
        facet_col="Ano", barmode="group",
        color_discrete_map={"Com apoio": "#2ca02c", "Sem apoio": "#d62728"},
        text=df_apoio["Taxa"].map(lambda v: f"{v:.1%}"),
        title="EvasÃ£o por Apoio Financeiro (FIES / ProUni) â 2023 vs 2024",
        labels={"Taxa": "Taxa MÃ©dia de EvasÃ£o"},
    )
    fig.update_traces(textposition="outside", textfont_size=12)
    fig.update_layout(
        **LAYOUT_PADRAO,
        yaxis_tickformat=".0%",
        yaxis_range=[0, 0.70],
    )
    fig.show()

grafico_apoio_financeiro()


# -----------------------------------------------------------------------
# GRÃFICO 5: Perfil DemogrÃ¡fico â Sexo
# -----------------------------------------------------------------------
def grafico_demografico_sexo():
    fig = go.Figure()

    # 2023
    if "PREDOM_SEXO" in df_completo_2023.columns:
        dados_2023 = df_completo_2023.groupby("PREDOM_SEXO")["TAXA_EVASAO"].mean().reset_index()
        dados_2023.columns = ["Grupo", "Taxa"]
        fig.add_trace(go.Bar(
            name="2023", x=dados_2023["Grupo"], y=dados_2023["Taxa"],
            marker_color=COR_2023,
            text=[f"{v:.1%}" for v in dados_2023["Taxa"]],
            textposition="outside"
        ))

    # 2024
    if "PREDOM_SEXO" in df_completo_2024.columns:
        dados_2024 = df_completo_2024.groupby("PREDOM_SEXO")["TAXA_EVASAO"].mean().reset_index()
        dados_2024.columns = ["Grupo", "Taxa"]
        fig.add_trace(go.Bar(
            name="2024", x=dados_2024["Grupo"], y=dados_2024["Taxa"],
            marker_color=COR_2024,
            text=[f"{v:.1%}" for v in dados_2024["Taxa"]],
            textposition="outside"
        ))

    fig.update_layout(
        **LAYOUT_PADRAO,
        title="EvasÃ£o por PredominÃ¢ncia de Sexo â 2023 vs 2024",
        xaxis_title="",
        yaxis_title="Taxa MÃ©dia de EvasÃ£o",
        yaxis_tickformat=".0%",
        yaxis_range=[0, 0.70],
        barmode="group",
    )
    fig.show()

grafico_demografico_sexo()


# -----------------------------------------------------------------------
# GRÃFICO 6 (CORRIGIDO): DistribuiÃ§Ã£o da Taxa de EvasÃ£o (Boxplot + Histograma)
# -----------------------------------------------------------------------

def grafico_distribuicao_corrigido():
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # GrÃ¡fico 1: Boxplot comparativo
    dados_box = [df_completo_2023['TAXA_EVASAO'].dropna(),
                 df_completo_2024['TAXA_EVASAO'].dropna()]

    bp = axes[0].boxplot(dados_box, labels=['2023', '2024'], patch_artist=True,
                          boxprops=dict(linewidth=1.5),
                          medianprops=dict(linewidth=2, color='darkred'),
                          whiskerprops=dict(linewidth=1.5),
                          capprops=dict(linewidth=1.5))

    # Cores das caixas
    bp['boxes'][0].set_facecolor(COR_2023)
    bp['boxes'][1].set_facecolor(COR_2024)

    axes[0].set_ylabel('Taxa de EvasÃ£o', fontsize=12)
    axes[0].set_title('DistribuiÃ§Ã£o da Taxa de EvasÃ£o - Boxplot', fontsize=14)
    axes[0].set_ylim(0, 1)
    axes[0].grid(axis='y', alpha=0.3)

    # Adicionar anotaÃ§Ãµes com mÃ©dias
    media_2023 = df_completo_2023['TAXA_EVASAO'].mean()
    media_2024 = df_completo_2024['TAXA_EVASAO'].mean()
    axes[0].text(1, media_2023 + 0.02, f'MÃ©dia 2023: {media_2023:.1%}', ha='center', fontsize=10)
    axes[0].text(2, media_2024 + 0.02, f'MÃ©dia 2024: {media_2024:.1%}', ha='center', fontsize=10)

    # GrÃ¡fico 2: Histogramas sobrepostos
    axes[1].hist(df_completo_2023['TAXA_EVASAO'].dropna(), bins=30, alpha=0.5,
                 label='2023', color=COR_2023, density=True)
    axes[1].hist(df_completo_2024['TAXA_EVASAO'].dropna(), bins=30, alpha=0.5,
                 label='2024', color=COR_2024, density=True)
    axes[1].set_xlabel('Taxa de EvasÃ£o', fontsize=12)
    axes[1].set_ylabel('Densidade', fontsize=12)
    axes[1].set_title('DistribuiÃ§Ã£o da Taxa de EvasÃ£o - Histograma', fontsize=14)
    axes[1].legend()
    axes[1].grid(axis='y', alpha=0.3)

    plt.suptitle('ComparaÃ§Ã£o da DistribuiÃ§Ã£o da EvasÃ£o (2023 vs 2024)', fontsize=16)
    plt.tight_layout()
    plt.show()

    # EstatÃ­sticas resumo
    print("\n--- EstatÃ­sticas Resumo da DistribuiÃ§Ã£o ---")
    print(f"{'MÃ©trica':<20} {'2023':>12} {'2024':>12}")
    print("-" * 45)
    print(f"{'MÃ©dia':<20} {df_completo_2023['TAXA_EVASAO'].mean():>11.1%} {df_completo_2024['TAXA_EVASAO'].mean():>11.1%}")
    print(f"{'Mediana':<20} {df_completo_2023['TAXA_EVASAO'].median():>11.1%} {df_completo_2024['TAXA_EVASAO'].median():>11.1%}")
    print(f"{'Desvio PadrÃ£o':<20} {df_completo_2023['TAXA_EVASAO'].std():>11.3f} {df_completo_2024['TAXA_EVASAO'].std():>11.3f}")
    print(f"{'Assimetria':<20} {df_completo_2023['TAXA_EVASAO'].skew():>11.3f} {df_completo_2024['TAXA_EVASAO'].skew():>11.3f}")

grafico_distribuicao_corrigido()

print("\nâ Todos os grÃ¡ficos gerados com sucesso!")
```

###-->dados agregados

```python
# ============================================================
# GERAR AGREGADOS PARA O STREAMLIT â rodar apÃ³s bloco 9
# ============================================================
import os
os.makedirs("streamlit_data", exist_ok=True)

mapa_regiao = {1: "Norte", 2: "Nordeste", 3: "Centro-Oeste",
               4: "Sudeste", 5: "Sul"}
mapa_org    = {1: "Universidade", 2: "Centro Univ.", 3: "Faculdade",
               4: "IF / CEFET", 5: "Univ. Especializada"}

def gerar_agregados(df, ano):
    resultados = {}

    # KPIs gerais
    resultados["kpi"] = pd.DataFrame([{
        "ano": ano,
        "n_cursos": len(df),
        "taxa_media": df["TAXA_EVASAO"].mean(),
        "taxa_mediana": df["TAXA_EVASAO"].median(),
    }])

    # Por rede
    df_rede = df.copy()
    df_rede["Rede"] = df_rede["TP_REDE_y"].map({1: "PÃºblica", 2: "Privada"})
    resultados["rede"] = (
        df_rede.groupby("Rede")["TAXA_EVASAO"]
        .agg(taxa_media="mean", n_cursos="count")
        .reset_index()
        .assign(ano=ano)
    )

    # Por regiÃ£o
    df_reg = df.copy()
    df_reg["RegiÃ£o"] = df_reg["CO_REGIAO"].map(mapa_regiao)
    resultados["regiao"] = (
        df_reg.dropna(subset=["RegiÃ£o"])
        .groupby("RegiÃ£o")["TAXA_EVASAO"]
        .agg(taxa_media="mean", n_cursos="count")
        .reset_index()
        .assign(ano=ano)
    )

    # Por organizaÃ§Ã£o acadÃªmica
    col_org = next((c for c in ["TP_ORGANIZACAO_ACADEMICA_x", "TP_ORGANIZACAO_ACADEMICA"]
                    if c in df.columns), None)
    if col_org:
        df_org = df.copy()
        df_org["OrganizaÃ§Ã£o"] = df_org[col_org].map(mapa_org)
        resultados["org"] = (
            df_org.dropna(subset=["OrganizaÃ§Ã£o"])
            .groupby("OrganizaÃ§Ã£o")["TAXA_EVASAO"]
            .agg(taxa_media="mean", n_cursos="count")
            .reset_index()
            .assign(ano=ano)
        )

    # Apoio financeiro
    rows_apoio = []
    for col_qt, label in [("QT_ING_FIES", "FIES"),
                           ("QT_ING_PROUNI_PARCIAL", "ProUni")]:
        if col_qt in df.columns:
            rows_apoio.append({"Programa": label, "SituaÃ§Ã£o": "Com apoio",
                                "taxa_media": df[df[col_qt] > 0]["TAXA_EVASAO"].mean(),
                                "n_cursos": (df[col_qt] > 0).sum(), "ano": ano})
            rows_apoio.append({"Programa": label, "SituaÃ§Ã£o": "Sem apoio",
                                "taxa_media": df[df[col_qt] == 0]["TAXA_EVASAO"].mean(),
                                "n_cursos": (df[col_qt] == 0).sum(), "ano": ano})
    if rows_apoio:
        resultados["apoio"] = pd.DataFrame(rows_apoio)

    # Perfil demogrÃ¡fico â sexo
    if "PREDOM_SEXO" in df.columns:
        resultados["sexo"] = (
            df.groupby("PREDOM_SEXO")["TAXA_EVASAO"]
            .agg(taxa_media="mean", n_cursos="count")
            .reset_index()
            .assign(ano=ano)
        )

    # DistribuiÃ§Ã£o completa (amostra para violin â mÃ¡x 5000 linhas)
    resultados["distribuicao"] = (
        df[["TAXA_EVASAO"]].dropna()
        .sample(min(5000, len(df)), random_state=42)
        .assign(ano=ano)
    )

    return resultados

# Gerar para os dois anos
agg_2023 = gerar_agregados(df_completo_2023, 2023)
agg_2024 = gerar_agregados(df_completo_2024, 2024)

# Combinar e salvar cada tema
temas = ["kpi", "rede", "regiao", "org", "apoio", "distribuicao"]
temas_com_sexo = temas + ["sexo"]

for tema in temas_com_sexo:
    partes = []
    if tema in agg_2023: partes.append(agg_2023[tema])
    if tema in agg_2024: partes.append(agg_2024[tema])
    if partes:
        df_final = pd.concat(partes, ignore_index=True)
        df_final.to_csv(f"streamlit_data/{tema}.csv", index=False)
        print(f"â {tema}.csv â {len(df_final)} linhas, "
              f"{df_final.memory_usage(deep=True).sum() / 1024:.1f} KB")

print("\nPasta streamlit_data pronta para subir no GitHub!")
```

```python
# ============================================================
# COLE ESTE BLOCO NO COLAB â apÃ³s o bloco 9
# Gera o kpi_v2.csv com contagens distintas de IES e cursos
# ============================================================

import os, pandas as pd
os.makedirs("streamlit_data", exist_ok=True)

def gerar_kpi_v2(df, ano):
    # Identificar colunas disponÃ­veis
    col_ies   = next((c for c in ["CO_IES", "co_ies"] if c in df.columns), None)
    col_curso = next((c for c in ["CO_CURSO", "co_curso", "NO_CURSO"] if c in df.columns), None)

    n_ies    = df[col_ies].nunique()   if col_ies   else None
    n_cursos = df[col_curso].nunique() if col_curso else None

    print(f"\nAno {ano}:")
    print(f"  VÃ­nculos totais  : {len(df):,}")
    print(f"  IES distintas    : {n_ies:,}"    if n_ies    else "  IES distintas    : coluna nÃ£o encontrada")
    print(f"  Cursos distintos : {n_cursos:,}" if n_cursos else "  Cursos distintos : coluna nÃ£o encontrada")
    print(f"  Taxa mÃ©dia       : {df['TAXA_EVASAO'].mean():.1%}")
    print(f"  Colunas disponÃ­veis: {list(df.columns)}")

    return pd.DataFrame([{
        "ano"              : ano,
        "n_vinculos"       : len(df),
        "n_instituicoes"   : n_ies,
        "n_cursos_distintos": n_cursos,
        "taxa_media"       : df["TAXA_EVASAO"].mean(),
        "taxa_mediana"     : df["TAXA_EVASAO"].median(),
    }])

kpi23 = gerar_kpi_v2(df_completo_2023, 2023)
kpi24 = gerar_kpi_v2(df_completo_2024, 2024)

kpi_v2 = pd.concat([kpi23, kpi24], ignore_index=True)
kpi_v2.to_csv("streamlit_data/kpi_v2.csv", index=False)

print("\nâ kpi_v2.csv salvo em streamlit_data/")
print(kpi_v2.to_string(index=False))
```

```python
# ============================================================
# COLE ESTE BLOCO NO COLAB â apÃ³s o bloco 9
# Gera dois CSVs de distribuiÃ§Ã£o para a aba de DistribuiÃ§Ã£o
# ============================================================

import os, pandas as pd, numpy as np
os.makedirs("streamlit_data", exist_ok=True)

def gerar_distribuicao(df, ano):

    taxa = df["TAXA_EVASAO"].dropna()
    taxa = taxa[(taxa >= 0) & (taxa <= 1)]

    # ââ Faixas de 10% em 10% ââââââââââââââââââââââââââââââ
    bins_10   = np.arange(0, 1.1, 0.10)
    labels_10 = [f"{int(b*100)}â{int((b+0.10)*100)}%" for b in bins_10[:-1]]
    faixas_10 = pd.cut(taxa, bins=bins_10, labels=labels_10,
                       include_lowest=True, right=False)
    df_10 = (faixas_10.value_counts().reindex(labels_10, fill_value=0)
                       .reset_index())
    df_10.columns = ["Faixa", "n_cursos"]
    df_10["pct_cursos"] = df_10["n_cursos"] / df_10["n_cursos"].sum()
    df_10["ano"]        = ano
    df_10["tipo"]       = "10pct"

    # ââ Faixas categÃ³ricas (25%) ââââââââââââââââââââââââââ
    bins_cat   = [0, 0.25, 0.50, 0.75, 1.01]
    labels_cat = ["Baixa (0â25%)", "MÃ©dia (25â50%)",
                  "Alta (50â75%)", "CrÃ­tica (75â100%)"]
    faixas_cat = pd.cut(taxa, bins=bins_cat, labels=labels_cat,
                        include_lowest=True, right=False)
    df_cat = (faixas_cat.value_counts().reindex(labels_cat, fill_value=0)
                         .reset_index())
    df_cat.columns = ["Faixa", "n_cursos"]
    df_cat["pct_cursos"] = df_cat["n_cursos"] / df_cat["n_cursos"].sum()
    df_cat["ano"]        = ano
    df_cat["tipo"]       = "categorias"

    # Print resumo
    print(f"\nAno {ano} â DistribuiÃ§Ã£o 10%:")
    for _, r in df_10.iterrows():
        print(f"  {r['Faixa']:>12} : {r['n_cursos']:>7,} cursos  ({r['pct_cursos']:.1%})")
    print(f"\nAno {ano} â Categorias:")
    for _, r in df_cat.iterrows():
        print(f"  {r['Faixa']:>22} : {r['n_cursos']:>7,} cursos  ({r['pct_cursos']:.1%})")

    return pd.concat([df_10, df_cat], ignore_index=True)

dist23 = gerar_distribuicao(df_completo_2023, 2023)
dist24 = gerar_distribuicao(df_completo_2024, 2024)
dist_v2 = pd.concat([dist23, dist24], ignore_index=True)

dist_v2.to_csv("streamlit_data/dist_faixas.csv", index=False)
print("\nâ dist_faixas.csv salvo em streamlit_data/")
print(f"   Linhas: {len(dist_v2)}")
```

```python
# Gera genero.csv - taxa de evasao por genero predominante
import os, pandas as pd
os.makedirs("streamlit_data", exist_ok=True)

mapa_genero = {
    "Maioria Mulheres": "Feminino predominante",
    "Maioria Homens":   "Masculino predominante",
    "Equilibrado":      "Equilibrado",
}

def gerar_genero(df, ano):
    df_g = df.copy()
    df_g["Genero"] = df_g["PREDOM_SEXO"].map(mapa_genero)
    return (
        df_g.dropna(subset=["Genero"])
        .groupby("Genero")["TAXA_EVASAO"]
        .agg(taxa_media="mean", n_cursos="count")
        .reset_index()
        .rename(columns={"Genero": "Genero predominante"})
        .assign(ano=ano)
    )

gen23 = gerar_genero(df_completo_2023, 2023)
gen24 = gerar_genero(df_completo_2024, 2024)
genero_df = pd.concat([gen23, gen24], ignore_index=True)
genero_df.to_csv("streamlit_data/genero.csv", index=False)
print("genero.csv salvo")
print(genero_df.to_string(index=False))
```

```python
# Gera grau.csv - taxa de evasao por grau academico
mapa_grau = {1: "Bacharelado", 2: "Licenciatura", 3: "Tecnologo"}

def gerar_grau(df, ano):
    df_g = df.copy()
    df_g["Grau"] = df_g["TP_GRAU_ACADEMICO"].map(mapa_grau)
    return (
        df_g.dropna(subset=["Grau"])
        .groupby("Grau")["TAXA_EVASAO"]
        .agg(taxa_media="mean", n_cursos="count")
        .reset_index()
        .assign(ano=ano)
    )

grau23 = gerar_grau(df_completo_2023, 2023)
grau24 = gerar_grau(df_completo_2024, 2024)
grau_df = pd.concat([grau23, grau24], ignore_index=True)
grau_df.to_csv("streamlit_data/grau.csv", index=False)
print("grau.csv salvo")
print(grau_df.to_string(index=False))
```

```python
# Atualiza apoio.csv - FIES e ProUni (integral + parcial)
def gerar_apoio_v2(df, ano):
    df = df.copy()
    if "QT_ING_PROUNI_TOTAL" not in df.columns:
        col_i = "QT_ING_PROUNII" if "QT_ING_PROUNII" in df.columns else None
        col_p = "QT_ING_PROUNIP" if "QT_ING_PROUNIP" in df.columns else None
        if col_i and col_p:
            df["QT_ING_PROUNI_TOTAL"] = df[col_i].fillna(0) + df[col_p].fillna(0)
        elif col_i:
            df["QT_ING_PROUNI_TOTAL"] = df[col_i].fillna(0)
        elif col_p:
            df["QT_ING_PROUNI_TOTAL"] = df[col_p].fillna(0)
    rows = []
    for col_qt, label in [("QT_ING_FIES", "FIES"), ("QT_ING_PROUNI_TOTAL", "ProUni")]:
        if col_qt not in df.columns:
            print(f"Coluna {col_qt} nao encontrada")
            continue
        rows.append({"Programa": label, "Situacao": "Com apoio",
                     "taxa_media": df[df[col_qt] > 0]["TAXA_EVASAO"].mean(),
                     "n_cursos": int((df[col_qt] > 0).sum()), "ano": ano})
        rows.append({"Programa": label, "Situacao": "Sem apoio",
                     "taxa_media": df[df[col_qt] == 0]["TAXA_EVASAO"].mean(),
                     "n_cursos": int((df[col_qt] == 0).sum()), "ano": ano})
    return pd.DataFrame(rows)

apoio23 = gerar_apoio_v2(df_completo_2023, 2023)
apoio24 = gerar_apoio_v2(df_completo_2024, 2024)
apoio_df = pd.concat([apoio23, apoio24], ignore_index=True)
apoio_df.to_csv("streamlit_data/apoio.csv", index=False)
print("apoio.csv atualizado (FIES + ProUni)")
print(apoio_df.to_string(index=False))
```

```python
# Gera top_cursos.csv - top 15 cursos com maior evasao por ano
# Minimo 30 ingressantes para evitar distorcoes estatisticas
col_rede = "TP_REDE_y" if "TP_REDE_y" in df_completo_2023.columns else "TP_REDE_x"
col_ies  = next((c for c in ["NO_IES", "SG_IES"] if c in df_completo_2023.columns), None)
mapa_rede_top = {1: "Publica", 2: "Privada"}
mapa_grau_top = {1: "Bacharelado", 2: "Licenciatura", 3: "Tecnologo"}

def gerar_top_cursos(df, ano, top_n=15, min_ing=30):
    df_t = df[df["QT_ING"] >= min_ing].copy()
    df_t["Rede"] = df_t[col_rede].map(mapa_rede_top)
    df_t["Grau"] = df_t["TP_GRAU_ACADEMICO"].map(mapa_grau_top)
    cols = ["NO_CURSO", "Rede", "Grau", "NO_REGIAO_IES", "TAXA_EVASAO", "QT_ING"]
    if col_ies:
        cols = ["NO_CURSO", col_ies] + cols[1:]
    df_top = df_t[cols].sort_values("TAXA_EVASAO", ascending=False).head(top_n).copy()
    if col_ies and col_ies != "NO_IES":
        df_top = df_top.rename(columns={col_ies: "NO_IES"})
    elif not col_ies:
        df_top.insert(1, "NO_IES", "desconhecido")
    df_top["ano"] = ano
    return df_top.reset_index(drop=True)

top23 = gerar_top_cursos(df_completo_2023, 2023)
top24 = gerar_top_cursos(df_completo_2024, 2024)
top_df = pd.concat([top23, top24], ignore_index=True)
top_df.to_csv("streamlit_data/top_cursos.csv", index=False)
print("top_cursos.csv salvo")
print(top_df[["NO_CURSO", "Rede", "Grau", "TAXA_EVASAO", "ano"]].to_string(index=False))
```
